```python
"""Visualize the LangGraph workflow for the Agentic Resume Optimizer."""

from pipeline import _graph

def main():
    # ASCII — always works, no extra dependencies
    print("\n=== Agentic Resume Optimizer — LangGraph Workflow ===\n")
    print(_graph.get_graph().draw_ascii())

    # PNG — saves to disk
    try:
        png_bytes = _graph.get_graph().draw_mermaid_png()
        with open("workflow.png", "wb") as f:
            f.write(png_bytes)
        print("\nFlowchart saved to workflow.png")
    except Exception as e:
        print(f"\nPNG render skipped: {e}")
        print("Install grandalf to enable PNG: pip install grandalf")

if __name__ == "__main__":
    main()
```
